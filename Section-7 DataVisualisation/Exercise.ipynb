```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x=np.linspace(0,5,11)
y=x**3
plt.plot(x,y)
plt.xlabel('countries')
plt.ylabel('population')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.35,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(8,2))
# for current_axes in axes:
#     current_axes.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title("first_plot")
axes[1].plot(y,x)
axes[1].set_title("second_plot")
plt.tight_layout()
```

```python
fig.savefig('my_picture.jpg')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X square",color="purple",linewidth=2)
ax.plot(x,x**5,label="X",lw=3,marker='o',markersize=10,markerfacecolor='red',markeredgewidth=2,markeredgecolor='green')
ax.plot(x,x**4,label="X Power4",linestyle='dashdot')
ax.legend()
ax.set_xlim([0,5])
ax.set_ylim([0,10])
```
